```python
#Question 2

loc
Label-based indexing: loc is used for label-based indexing. You can use row and column labels to select data.
Inclusive: When specifying a range, both the start and the end labels are included.
Slicing with labels: Allows you to use labels for slicing and provides more flexibility with named indexes.

iloc
Integer-based indexing: iloc is used for integer-based (position-based) indexing. You use row and column indices to select data.
Exclusive: When specifying a range, the end index is excluded.
Slicing with integers: Only integer positions can be used for slicing.
```

```python
#Question 6

n Pandas, window functions are operations that perform calculations across a specified subset (or window) of data, typically defined by a window size or based on specific criteria like time periods or indices. These functions are extremely useful in performing rolling or expanding calculations over data, especially in time series analysis or other sequential data contexts.

Types of Window Functions in Pandas:
Rolling Window Functions:

.rolling(): Computes metrics over a rolling window of a specified size.
Examples: rolling_mean(), rolling_sum(), rolling_std().
Expanding Window Functions:

.expanding(): Computes metrics across all previous data points.
Examples: expanding_mean(), expanding_sum().
Exponential Moving Window Functions:

.ewm(): Applies exponential weights to data, giving more weight to recent observations.
Examples: ewm_mean(), ewm_std().
```

```python
#Question 9 

import pandas as pd


def convert_and_sort_categorical_data(file_path, column_name, category_order):
    try:
     
        df = pd.read_csv(file_path)
        
        
        df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
        
        
        sorted_df = df.sort_values(by=column_name)
        
      
        print("Sorted Data:")
        print(sorted_df)
        
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except KeyError:
        print(f"Error: Column '{column_name}' not found in the CSV file.")


file_path = input("Enter the file path of the CSV file: ")
column_name = input("Enter the name of the column to convert to categorical: ")
category_order = input("Enter the category order (comma-separated): ").split(',')


convert_and_sort_categorical_data(file_path, column_name, category_order)
```

```python
#Question 10

import pandas as pd
import matplotlib.pyplot as plt


def visualize_sales_data(file_path):
    try:
        
        df = pd.read_csv(file_path)
        
        
       
        df['Date'] = pd.to_datetime(df['Date'])
        
        
        sales_pivot = df.pivot_table(index='Date', columns='Product Category', values='Sales', aggfunc='sum', fill_value=0)
        
      
        sales_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
        
        plt.title('Sales of Product Categories Over Time')
        plt.xlabel('Date')
        plt.ylabel('Sales')
        plt.xticks(rotation=45)
        plt.legend(title='Product Category', bbox_to_anchor=(1, 1))
        plt.tight_layout()
        
       
        plt.show()
        
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")


file_path = input("Enter the file path of the CSV file containing sales data: ")


visualize_sales_data(file_path)
```

```python
#Question 11

import pandas as pd
from tabulate import tabulate

# Function to calculate mean, median, and mode of test scores
def calculate_statistics(file_path):
    try:
        # Read CSV file into a Pandas DataFrame
        df = pd.read_csv(file_path)
        
        # Calculate mean, median, and mode of test scores
        mean_score = df['Test Score'].mean()
        median_score = df['Test Score'].median()
        mode_scores = df['Test Score'].mode().tolist()  # mode() returns a Series, convert to list
        
        # Prepare results in a tabular format
        results = [
            ["Mean", mean_score],
            ["Median", median_score],
            ["Mode", ", ".join(map(str, mode_scores))] 
        ]
        
      
        print(tabulate(results, headers=["Statistic", "Value"]))
        
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except KeyError:
        print(f"Error: CSV file '{file_path}' does not contain expected columns.")

file_path = input("Enter the file path of the CSV file containing the student data: ")


calculate_statistics(file_path)
```
